```python
######### PROVIDING LABELS ##############

# labeling the axes

import pylab as plt

mySamples = []
myLinear = []
myQuadratic = []
myCubic = []
myExponential = []

for i in range(0, 30):
    mySamples.append(i)
    myLinear.append(i)
    myQuadratic.append(i**2)
    myCubic.append(i**3)
    myExponential.append(1.5**i)

plt.figure('lin')
plt.clf()   # cleaning the window
plt.title('Linear')  # Adding title
# Adding label for Linear (MUST MAKE FIGURE Active before labeling)
plt.xlabel('Sample Points')
plt.ylabel('Linear Function')
plt.plot(mySamples, myLinear)

plt.figure('quad')
plt.clf()   # cleaning the window
plt.title('Quadratic')  # Adding title
plt.xlabel('Test Points')
plt.ylabel('Quadratic Function')
plt.plot(mySamples, myQuadratic)

plt.figure('cube')
plt.clf()   # cleaning the window
plt.title('Cubic')  # Adding title
plt.xlabel('Test2 Points')
plt.ylabel('Cubic Function')
plt.plot(mySamples, myCubic)

plt.figure('expo')
plt.clf()   # cleaning the window
plt.title('Exponential')  # Adding title
plt.xlabel('Expo Points')
plt.ylabel('Exponential Function')
plt.plot(mySamples, myExponential)

plt.show()

#### CLEANING UP WINDOWS #######
#    - command for it is:  plt.clf()
```
